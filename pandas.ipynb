```python
import pandas as pd

# Sample data
data = {
    'W': [2, 6, 2, 16, 30],
    'X': [79, -29, 21, -1, 49],
    'Y': [-8, 88, -26, 3, -48],
    'Z': [-86, -80, -13, 51, -99]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])

def filter_dataframe(df, filters):
    # Create a mask with all True values (include all rows initially)
    mask = pd.Series([True] * len(df), index=df.index)
    
    # Apply filters based on the provided dictionary
    for column, values in filters.items():
        if column in df.columns:
            mask &= df[column].isin(values)
    
    # Return the filtered DataFrame
    return df[mask]

# Test the function
filter_criteria = {
    'W': [2, 16],
    'X': [79, -1]
}
filtered_df = filter_dataframe(df, filter_criteria)
print(filtered_df)
```

```python
# Define filter criteria
filter_criteria1 = {
    'W': [2, 16],
    'X': [79, -1]
}
filtered_df1 = filter_dataframe(df, filter_criteria1)
print(filtered_df1)

# Define another set of filter criteria
filter_criteria2 = {
    'Y': [-8, -26, 3],
    'Z': [-86, 51]
}
filtered_df2 = filter_dataframe(df, filter_criteria2)
print(filtered_df2)

# Define a filter for a single column
filter_criteria3 = {
    'W': [2, 6, 30]
}
filtered_df3 = filter_dataframe(df, filter_criteria3)
print(filtered_df3)
```

```python
import pandas as pd

# Sample data
data = {
    'W': [2, 6, 2, 16, 30],
    'X': [79, -29, 21, -1, 49],
    'Y': [-8, 88, -26, 3, -48],
    'Z': [-86, -80, -13, 51, -99]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])

def filter_dataframe(df, filters):
    # Create a mask with all True values (include all rows initially)
    mask = pd.Series([True] * len(df), index=df.index)
    
    # Apply filters based on the provided dictionary
    for column, values in filters.items():
        if column in df.columns:
            mask &= df[column].isin(values)
    
    # Return the filtered DataFrame
    return df[mask]

# Define filter values using variables
W_values = [2, 16]
X_values = [79, -1]

# Construct the filter criteria dictionary using the variables
filter_criteria = {
    'W': W_values,
    'X': X_values
}

# Test the function
filtered_df = filter_dataframe(df, filter_criteria)
print(filtered_df)
```

```python
import pandas as pd

# Sample data
data = {
    'W': [2, 6, 2, 16, 30],
    'X': [79, -29, 21, -1, 49],
    'Y': [-8, 88, -26, 3, -48],
    'Z': [-86, -80, -13, 51, -99]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])
```

```python
# Create a pivot table
pivot_table = pd.pivot_table(df, values=['X', 'Y', 'Z'], index=['W'], aggfunc='mean')

print(pivot_table)
```

```python
pivot_table_sum = pd.pivot_table(df, values=['X', 'Y', 'Z'], index=['W'], aggfunc='sum')
print(pivot_table_sum)
```

```python
pivot_table_multiple = pd.pivot_table(df, values=['X', 'Y', 'Z'], index=['W'], aggfunc=[sum, len])
print(pivot_table_multiple)
```

```python
pivot_table_multiple = pd.pivot_table(df, values=['X', 'Y', 'Z'], index=['W'], aggfunc=[sum, 'mean'])
print(pivot_table_multiple)
```

```python
import json

# JSON input as a string
json_input = '''
[
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35}
]
'''

# Parse JSON input and convert to Python list
python_list = json.loads(json_input)

# Initialize separate lists for Alice, Bob, and Charlie
alice_list = []
bob_list = []
charlie_list = []

# Iterate through the list and separate the data
for item in python_list:
    if item['name'] == 'Alice':
        alice_list.append(item)
    elif item['name'] == 'Bob':
        bob_list.append(item)
    elif item['name'] == 'Charlie':
        charlie_list.append(item)

# Print the resulting lists
print("Alice's list:", alice_list)
print("Bob's list:", bob_list)
print("Charlie's list:", charlie_list)
```


Sure, let's create a dummy REST API using Flask that accepts JSON input and returns JSON output. We'll set up a simple Flask application that listens for POST requests with JSON data and responds with a JSON output.

Step-by-Step Guide
Install Flask: Ensure Flask is installed. If not, install it using pip.
Set Up Flask Application: Create a Flask application with an endpoint to handle POST requests.
Handle JSON Input and Output: Parse the input JSON and return a JSON response.

Step 2: Set Up Flask Application
Create a new Python file, e.g., app.py, and set up the Flask application:

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/process', methods=['POST'])
def process_json():
    try:
        # Parse the JSON input
        data = request.get_json()

        # Assuming the JSON input is a list of dictionaries
        names_list = []
        ages_list = []

        for item in data:
            names_list.append(item['name'])
            ages_list.append(item['age'])

        # Create response JSON
        response = {
            'names': names_list,
            'ages': ages_list
        }

        return jsonify(response), 200
    except Exception as e:
        return jsonify({'error': str(e)}), 400

if __name__ == '__main__':
    app.run(debug=True)
```

Explanation
Import Flask and Required Modules: Import the Flask module and other required functions.
Create Flask App: Initialize the Flask app.
Define Route: Define a route /process that accepts POST requests.
Process JSON Input: Inside the route function, parse the JSON input, create separate lists for names and ages, and construct the JSON response.
Error Handling: Handle any potential errors and return appropriate error messages.
Run the App: Run the Flask application.
Step 3: Run the Flask Application
To run the Flask application, execute the Python script:

```python
#in shell
python app.py
```
